```python
import matplotlib.pyplot as plt
import pandas as pd

# Membaca file CSV
dataframe = pd.read_csv('stock - Copy.csv')

# Mendapatkan kolom Purchase Price, Current Price, dan Gain/Loss
purchase_price = dataframe['Purchase Price']
current_price = dataframe['Current Price']
gain_loss = dataframe['Gain/Loss']

# Mengubah ukuran gambar histogram
plt.figure(figsize=(10, 8))

# Membuat histogram untuk Purchase Price
plt.hist(purchase_price, bins=15, edgecolor='black')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')
plt.show()

# Mengubah ukuran gambar histogram
plt.figure(figsize=(10, 8))

# Membuat histogram untuk Current Price
plt.hist(current_price, bins=15, edgecolor='black')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')
plt.show()

# Mengubah ukuran gambar histogram
plt.figure(figsize=(12, 10))

# Membuat histogram untuk Gain/Loss
plt.hist(gain_loss, bins=15, edgecolor='black')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')
plt.show()
```

```python
import pandas as pd

# Membaca file CSV yang berisi data saham
dataframe = pd.read_csv('stock - Copy.csv')

# Mengurutkan data berdasarkan kolom "Market Value" secara menurun
sorted_data = dataframe.sort_values(by='Dividend/share', ascending=False)

# Mengambil 5 baris pertama setelah diurutkan
top_5_data = sorted_data.head(5)

# Memilih kolom "Stock Name" dan "Symbol" dari data teratas
stock_names = top_5_data['Stock Name']
symbols = top_5_data['Symbol']

# Membuat kalimat untuk atribut "Stock Name" dan "Symbol"
sentences = []
for name, symbol in zip(stock_names, symbols):
    sentence = f"Stock Name: {name}, Symbol: {symbol}"
    sentences.append(sentence)

# Menampilkan kalimat
for sentence in sentences:
    print(sentence)
```
